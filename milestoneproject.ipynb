```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("meetinglistdetails_2026_05_30_2026_05_30.csv")
print("="*50)
print("DATASET OVERVIEW")
print("="*50)

print("\nFirst 5 Rows:")
print(df.head())

print("\nShape of Dataset:")
print(df.shape)

print("\nColumn Names:")
print(df.columns.tolist())

print("\nDataset Information:")
print(df.info())
print("\n" + "="*50)
print("MISSING VALUES")
print("="*50)

print(df.isnull().sum())


print("\n" + "="*50)
print("STATISTICAL SUMMARY")
print("="*50)

print(df.describe(include='all'))


print("\n" + "="*50)
print("MEETING STATISTICS")
print("="*50)

print("Total Meetings:", len(df))


if 'Participants' in df.columns:
    print("Total Participants:", df['Participants'].sum())


if 'Duration (minutes)' in df.columns:
    print("Average Duration:", round(df['Duration (minutes)'].mean(),2), "minutes")


if 'Duration (minutes)' in df.columns:
    print("Longest Meeting:", df['Duration (minutes)'].max(), "minutes")


if 'Topic' in df.columns:

    print("\nTop 10 Meeting Topics:")
    print(df['Topic'].value_counts().head(10))

    plt.figure(figsize=(10,5))
    df['Topic'].value_counts().head(10).plot(kind='bar')
    plt.title('Top 10 Meeting Topics')
    plt.xlabel('Topic')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


if 'Host name' in df.columns:

    print("\nTop 10 Hosts:")
    print(df['Host name'].value_counts().head(10))

    plt.figure(figsize=(10,5))
    df['Host name'].value_counts().head(10).plot(kind='bar')
    plt.title('Top 10 Hosts')
    plt.xlabel('Host Name')
    plt.ylabel('Number of Meetings')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

if 'Duration (minutes)' in df.columns:

    plt.figure(figsize=(8,5))
    df['Duration (minutes)'].hist(bins=20)
    plt.title('Meeting Duration Distribution')
    plt.xlabel('Duration (minutes)')
    plt.ylabel('Frequency')
    plt.show()

if 'Participants' in df.columns:

    plt.figure(figsize=(8,5))
    df['Participants'].hist(bins=20)
    plt.title('Participants Distribution')
    plt.xlabel('Participants')
    plt.ylabel('Frequency')
    plt.show()

df.to_csv("cleaned_meeting_data.csv", index=False)

print("\nAnalysis Completed Successfully!")
print("Cleaned dataset saved as: cleaned_meeting_data.csv")
```

```python
import pandas as pd

df = pd.read_csv("meetinglistdetails_2026_05_30_2026_05_30.csv")

while True:

    student = input("\nEnter Student Name: ")

    result = df[
        df["Name (original name)"]
        .astype(str)
        .str.lower()
        .str.contains(student.lower(), na=False)
    ]

    if len(result) > 0:

        print("\nSTATUS : PRESENT")

        print("\nDetails:")
        print(result[
            [
                "Name (original name)",
                "Join time",
                "Leave time"
            ]
        ])

    else:

        print("\nSTATUS : ABSENT")

    choice = input("\nCheck another student? (y/n): ")

    if choice.lower() != 'y':
        break
```
